## 3. Orbital Period vs. Radius for Transiting *Kepler* Planets

Now it's your turn! Make a plot of the transiting planets discovered by Kepler with orbital period on the x-axis and planet radius (in units of Earth radius) on the y-axis.

Be sure to label your axes with units! Refer to the header in the .csv file for more information.

```python
# path to .csv file
data_path = "file:///C:/Users/Ryan%20Chang/OneDrive/Documents/Intro2Astro_2024/data/kepler_transit_exoplanets_data.csv"

# ignore file header
data = pd.read_csv(data_path, comment='#')

# plots data with orbital period (days) on x axis and mass (earth masses) on y axis
plt.plot(data['pl_orbper'], data['pl_bmasse'], '.', alpha=0.3)

# labeling axes
plt.xlabel('Orbital Period [days]', fontsize=14)
plt.ylabel('Mass [M$_E$]', fontsize=14)

# logarithmic scales
plt.xscale('log')
plt.yscale('log')

plt.show()
```

I chose to plot right ascension and declination to see the area surveyed by the Kepler mission. The resulting plot reveals a cross-shaped area segmented into squares (which were probably invidual areas surveyed by Kepler), covering only a small portion of the celestial sphere. A slight distortion is also visible from plotting spherical coordinates onto a 2D figure. 

The small size of the search area really puts into perspective the vastness of the universe: Kepler has discovered so much, yet it barely scratches the surface of the plethora of exoplanets detectable from Earth, which in itself is an infinitesimal fraction of the number of exoplanets that exist in the universe.
